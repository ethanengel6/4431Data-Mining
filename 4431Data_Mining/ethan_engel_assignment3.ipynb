Iris Species Jupyter Notebook

Iris measurements were taken in centimeters (cm)

```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
from numpy import linalg

irisdf = datasets.load_iris(as_frame=True)
iris = irisdf.data
iris['species'] = irisdf.target.map(lambda x: irisdf.target_names[x])
# drop cm from column name
iris.rename(columns={'sepal width (cm)': 'sepal width', 'sepal length (cm)':'sepal length', 
                       'petal length (cm)':'petal length', 'petal width (cm)': 'petal width'}, inplace=True)
species_list = ['setosa', 'versicolor', 'virginica']

# EDA 
print ("Target Names: " ,irisdf.target_names)
#(first dimension = (ROWS) ie number of observations, second dimensions = (COLUMNS) ie number of features)
print("Shape of data: ", irisdf.data.shape)
# print features list
features = irisdf.data[:]
print(features)


# plot graph
fig, axs = plt.subplots(figsize=(10, 10))
scatter = plt.scatter(iris['sepal width'], iris['sepal length'], c=irisdf.target)
plt.xlabel('sepal width')
plt.ylabel('sepal length')
plt.legend(handles=scatter.legend_elements()[0], labels=species_list)
plt.title('Sepal Width vs Length')
```

Setosa is very seperated from the cluster, so would be easily classifiable with the given attribute data. Versicolor and Virginica are intermixed. When observed as a single data set, there is no discernible association.  However, within each individual category we see moderate positive correlations.

```python
# Add the petal length as a third dimension to the plot.

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(projection='3d')

ax.scatter(iris['sepal width'], iris['sepal length'], 
           iris['petal length'], c=irisdf.target)

ax.set_xlabel('sepal width')
ax.set_ylabel('sepal length')
ax.set_zlabel('petal length')

plt.title('Sepal Length(x), Width(y), Petal length(z)')
```

Adding the third dimension of petal length to the plot does not seem to shed any further light upon the classification scheme.

```python
np.dot(iris['sepal length'], iris['sepal width'])
```

```python
np.linalg.norm(iris['sepal length'])
```

```python
dis = np.linalg.norm(iris['sepal length'] - iris['sepal width'])
print(dis)
```

Conclusion for calculations: These metrics do not seem to be especially meaningful.

Overall conclusions about the different species is that the setosa is easily classifiable, in fact, completely separate from the other two species using the given attributes.  In order to be able to distinguish between versicolor & virginica, different attribute data would have to be gathered and added to the model(s) and/or new classification methods would need to be employed.
